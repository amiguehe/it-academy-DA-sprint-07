1 - Calculadora de l'índex de massa corporal
Input corregido para que se pueda introducir tanto "," como "." y delimitado a tres digitos en el resultado de IMC

```python
#ex1(con input)
weight = float(input("Ingrese su peso en kg: ").replace(",", "."))
height = float(input("Ingrese su altura en cm: ").replace(",", "."))

# Convert height from cm to meters
height /= 100

# Calculate BMI
bmi = weight / (height ** 2)

# Determine BMI category
print("altura",height)
print("peso",weight)
print ("IMC",'%.3f'%bmi)
if bmi < 18.5:
    print("Bajo peso")
elif 18.5 <= bmi < 24.9:
    print("Peso normal")
elif 24.9 <= bmi < 29.9:
    print("Sobrepeso")
else:
    print("Obesidad")
```

- Exercici 2
Convertidor de temperatures.

Formato por ahora no cambiado ya que la herramienta formateo crashea visual studio code.

```python
#ex_2

def convert_temperature(value, input_scale, output_scale):
    
    if input_scale == 'C':
        if output_scale == 'F':
            return value * 1.8 + 32
        elif output_scale == 'K':
            return value + 273.15
        elif output_scale =="N":
            return value / 3.03030303
        else:
            return value
    elif input_scale == 'F':
        if output_scale == 'C':
            return (value - 32) / 1.8
        elif output_scale == 'K':
            return (value + 459.67) * 5 / 9
        elif output_scale =="N":
            return (value - 32)*11/60
        else:
            return value
    elif input_scale == 'K':
        if output_scale == 'C':
            return value - 273.15
        elif output_scale == 'F':
            return value * 9 / 5 - 459.67
        elif output_scale =="N":
            return ( value - 273.15 ) * 33 / 100
        else:
            return value
    elif input_scale =="N":
        if output_scale =="C":
            return value * 3.03030303
        elif output_scale =="F":
            return (value * 60/11 + 32)
        elif output_scale =="K":
            return (value * 100/33 + 273.15)
        else:
            return value
    else:
        return value

while True:
    # Prompt the user for input
    print('Enter the input temperature value:')
    value = float(input())
    print('Enter the input temperature scale (C, F, K or N):')
    input_scale = input().upper()
    print('Enter the output temperature scale (C, F, K or N):')
    output_scale = input().upper()

    # Convert the temperature and print the result
    result = convert_temperature(value, input_scale, output_scale)
    print(f'{value} {input_scale} = {result} {output_scale}')

    # Prompt the user to continue or quit
    print('Enter q to quit, or any other key to continue:')
    choice = input()
    if choice.lower() == 'q':
        break
```

- Exercici 3
Comptador de paraules d'un text.

Escriu una funció que donat un text, mostri les vegades que apareix cada paraula.

```python
#ex3

import re
intro =input("escribe aqui tu frase: ")

wordlist = intro.split()

wordfreq = []
for w in wordlist:
    wordfreq.append(wordlist.count(w))

normal_string =re.sub("[^A-Z]", " ", intro,0,re.IGNORECASE)
print("String after conversion: ",normal_string)
print("String\n" + intro +"\n")
print("List\n" + str(wordlist) + "\n")
print("Frequencies\n" + str(wordfreq) + "\n")
print("Pairs\n" + str(list(zip(wordlist, wordfreq))))
```

Exercici 2
Conversió de tipus de dades

```python
#codigo simplificado:

lista = ['1.3', 'one', '1e10', 'seven', '3-1/2', ('2', 1, 1.4, 'not-a-number'), tuple([1, 2, '3', '3.4'])]

floats = []
non_floats = []

for item in lista:
    if isinstance(item, (int, float)):
        floats.append(item)
    elif isinstance(item, str):
        try:
            float_value = float(item)
            floats.append(float_value)
        except ValueError:
            non_floats.append(item)
    elif isinstance(item, tuple):
        for sub_item in item:
            try:
                if isinstance(sub_item, (int, float)):
                    floats.append(sub_item)
                else:
                    float_value = float(str(sub_item))
                    floats.append(float_value)
            except ValueError:
                non_floats.append(sub_item)


print("Floats:", floats)
print("Non-floats:", non_floats)
```

Exercici 1
Comptador i endreçador de paraules d'un text.
Importante: Poner la dirección adecuada donde esté el documento"tu_me_quieres_blanca.txt"

```python
import re

# Open the file
with open('F:\\tu_me_quieres_blanca.txt', 'r', encoding='utf-8') as file:
    # Read the file content
    content = file.read()

# Extract words using regular expressions
words = re.findall(r'\b[a-zA-ZñÑáéíóúÁÉÍÓÚüÜ.,;\'\"`´\-]+\b', content.lower(), flags=re.UNICODE)

# Create a dictionary to store word frequencies
word_freq = {}

# Count word frequencies and group them by starting letter
for word in words:
    first_letter = word[0]
    word_freq.setdefault(first_letter, {})
    word_freq[first_letter][word] = word_freq[first_letter].get(word, 0) + 1

# Sort the dictionary by keys (letters) alphabetically
sorted_word_freq = sorted(word_freq.items())

# Print the sorted word frequencies with titles for each letter
for letter, words_dict in sorted_word_freq:
    print(f'\nWords starting with letter {letter.upper()}:')
    for word, freq in sorted(words_dict.items()):
        print(f'{word}: {freq}')
```
